```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
data_df = pd.read_csv('Purchased.csv')

data_df.head()
```

```python
data_df['Salary'] = data_df['Salary'].fillna(value=63778)  #age 39
```

```python
data_df['Age'] = data_df['Age'].fillna(value=39)  #age 39
data_df
```

```python
k = 0
for i in data_df["Country"]:
    if i == "Germany":
        data_df["Country"].loc[k] = 1
    elif i == "Spain":
        data_df["Country"].loc[k] = 2
    else:
        data_df["Country"].loc[k] = 3
    k += 1
```

```python
data_df
```

```python
k = 0
for i in data_df["Purchased"]:
    if i == "Yes":
        data_df["Purchased"].loc[k] = 1
    elif i == "No":
        data_df["Purchased"].loc[k] = 0
    k += 1
```

```python
data_df
```

```python
data_df['Age'] = data_df['Age'] / max(data_df['Age'])
```

```python
data_df['Salary'] = data_df['Salary'] / max(data_df['Salary'])
```

```python
data_df['Country'] = data_df['Country'] / max(data_df['Country'])
```

```python
data_df[['Country', 'Age', 'Salary']].shape
```

```python
model = tf.keras.Sequential(
    [tf.keras.layers.Dense(units=1, input_shape=[3]), tf.keras.layers.Dense(256, activation='relu'),
     tf.keras.layers.Dense(128, activation='sigmoid'), tf.keras.layers.Dense(1, activation='softmax')])

model.summary()
```

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```
