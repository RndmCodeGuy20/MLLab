```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
sns.set_style('darkgrid')
colors = {
    "Amber": "#ffbe0b",
    "Orange Pantone": "#fb5607",
    "Winter Sky": "#ff006e",
    "Blue Violet": "#8338ec",
    "Azure": "#3a86ff"
}
```

```python
suv_df = pd.read_csv('./assets/suv_data.csv')

suv_df = suv_df.drop('User ID', axis=1)
```

```python
suv_df.head()
```

```python
plt.figure(figsize=(5, 5))

plt.scatter(suv_df["Age"].loc[suv_df["Purchased"] == 1], suv_df["EstimatedSalary"].loc[suv_df["Purchased"] == 1],
            color=colors["Amber"])
plt.scatter(suv_df["Age"].loc[suv_df["Purchased"] == 0], suv_df["EstimatedSalary"].loc[suv_df["Purchased"] == 0],
            color=colors["Azure"])

plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(suv_df[["Age", "EstimatedSalary"]], suv_df["Purchased"],
                                                    test_size=0.33, random_state=42)
```

```python
linear_svm = SVC(kernel='linear', C=1.0)
```

```python
linear_svm.fit(X_train, y_train)
```

```python
svm = SVC()
```

```python
svm.fit(X_train, y_train)
```

```python
svm.predict([[35, 20000]])
```

```python
accuracy_score(svm.predict(X_test), y_test) * 100
```
