```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')
colors = {
    "Amber": "#ffbe0b",
    "Orange Pantone": "#fb5607",
    "Winter Sky": "#ff006e",
    "Blue Violet": "#8338ec",
    "Azure": "#3a86ff"
}
```

```python
suv_df = pd.read_csv('./assets/suv_data.csv')

suv_df = suv_df.drop('User ID', axis=1)
```

```python
suv_df.head()
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(suv_df["Age"].loc[suv_df["Purchased"] == 1], suv_df["EstimatedSalary"].loc[suv_df["Purchased"] == 1],
            color=colors["Amber"])
plt.scatter(suv_df["Age"].loc[suv_df["Purchased"] == 0], suv_df["EstimatedSalary"].loc[suv_df["Purchased"] == 0],
            color=colors["Azure"])

plt.show()
```
